```r
# Load required package
library(ggplot2)

# Sample daily high and low temperatures for April 1–30
highs <- c(34, 35, 35, 35, 34, 35, 35, 35, 36, 36,
           36, 36, 36, 35, 37, 37, 37, 36, 36, 37,
           36, 37, 37, 37, 37, 36, 36, 37, 36, 36)

lows <- c(26, 27, 27, 28, 26, 26, 27, 28, 28, 27,
          28, 29, 28, 29, 29, 26, 25, 26, 28, 27,
          28, 28, 27, 27, 27, 26, 28, 29, 30, 28)

# Create data frame
weather_data <- data.frame(
  Day = 1:30,
  High = highs,
  Low = lows
)

# Plot histogram for Highs
ggplot(weather_data, aes(x = High)) +
  geom_histogram(binwidth = 1, fill = "orange", color = "black") +
  labs(title = "April 2025 - High Temperature Histogram",
       x = "High Temperature (°C)",
       y = "Frequency") +
  theme_minimal()

# Plot histogram for Lows
ggplot(weather_data, aes(x = Low)) +
  geom_histogram(binwidth = 1, fill = "lightblue", color = "black") +
  labs(title = "April 2025 - Low Temperature Histogram",
       x = "Low Temperature (°C)",
       y = "Frequency") +
  theme_minimal()-
```
